```python
%%sql
CREATE TABLE employee (
    emp_id INT PRIMARY KEY AUTO_INCREMENT,
    emp_name VARCHAR(50) NOT NULL,
    department VARCHAR(50),
    salary DECIMAL(10,2),
    hire_date DATE
);
```

```python
%%sql
INSERT INTO employee (emp_name, department, salary, hire_date)
VALUES
('Arun', 'IT', 60000, '2023-01-10'),
('Priya', 'HR', 45000, '2022-05-15'),
('Karthik', 'Finance', 75000, '2021-03-20'),
('Divya', 'IT', 52000, '2024-07-01');
```

```python
%%sql
ALTER TABLE employee
MODIFY salary DECIMAL(12,2);
```

```python
%%sql
DESC employee;
```

```python
%%sql
select * from employee;
```

```python
%%sql
INSERT INTO high_paid_employees (emp_id, emp_name, department, salary, hire_date)
SELECT emp_id, emp_name, department, salary, hire_date
FROM employee
WHERE salary > 50000;

```

```python
%%sql
select * from high_paid_employees;
```

```python
%%sql
CREATE TABLE rides (
    ride_id INT,
    driver_id INT,
    rider_id INT,
    pickup_location VARCHAR(100),
    dropoff_location VARCHAR(100),
    ride_date DATETIME,
    fare DECIMAL(10,2)
);
```
